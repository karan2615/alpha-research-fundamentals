## 1. Data Loading

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy.stats import mstats
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import RidgeCV, LassoCV
from sklearn.metrics import r2_score
from xgboost import XGBRegressor
import yfinance as yf

DATA = '../../seminar_02/homework/data/'

# ── Load final panel from HW2 ─────────────────────────────────────────
panel = pd.read_csv(DATA + 'panel_clean.csv', parse_dates=['date'])

# ── Load prices to compute new stock-specific signals ─────────────────
prices_clean = pd.read_csv(DATA + 'sp500_prices_clean.csv', index_col=0, parse_dates=True)

print(f'Panel shape:   {panel.shape}')
print(f'Prices shape:  {prices_clean.shape}')
print(f'Columns: {list(panel.columns)}')
```

## 2. Feature Engineering

Starting from the HW2 baseline (momentum + macro features), we add and evaluate stock-specific signals computable from price data alone.

All features are winsorized at the 1st/99th percentile and z-scored for comparability.

| Feature | Description | Expected Sign |
|---|---|---|
| reversal | 21-day simple return | Negative (mean reversion) |
| mom_3m | 3-month log return excl. last month | Positive |
| mom_6m | 6-month log return excl. last month | Positive |
| volatility | 21-day rolling std of daily returns | Negative (low-vol anomaly) |
| ratio_52w | Price / 52-week high | Tested — came out negative in OLS |

```python
# ── Compute all features ─────────────────────────────────────────────
# Short-term reversal
reversal = prices_clean.pct_change(21)

# Multi-horizon momentum
mom_3m = np.log(prices_clean).diff(63).shift(21)
mom_6m = np.log(prices_clean).diff(126).shift(21)

# Volatility
volatility = prices_clean.pct_change().rolling(21).std()

# 52-week high ratio
ratio_52w = prices_clean / prices_clean.rolling(252).max()

# ── Melt and merge into panel ─────────────────────────────────────────
def to_long(df, name):
    long = df.stack().reset_index()
    long.columns = ['date', 'ticker', name]
    return long

for df, name in [(reversal, 'reversal'), (mom_3m, 'mom_3m'), (mom_6m, 'mom_6m'),
                 (volatility, 'volatility'), (ratio_52w, 'ratio_52w')]:
    panel = panel.merge(to_long(df, name), on=['date', 'ticker'], how='left')

# Price acceleration
panel['price_accel'] = panel['mom_3m'] - panel['mom_6m']

panel = panel.dropna(subset=['reversal', 'mom_3m', 'mom_6m', 'volatility', 'ratio_52w'])

# ── Winsorize and standardize ─────────────────────────────────────────
new_features = ['reversal', 'mom_3m', 'mom_6m', 'volatility']
for col in new_features:
    panel[col] = mstats.winsorize(panel[col], limits=[0.01, 0.01])
    panel[col] = (panel[col] - panel[col].mean()) / panel[col].std()

# ratio_52w: bounded 0-1, just standardize
panel['ratio_52w'] = (panel['ratio_52w'] - panel['ratio_52w'].mean()) / panel['ratio_52w'].std()

print(f'Panel shape after feature engineering: {panel.shape}')
print(f'Columns: {list(panel.columns)}')
```

### Feature Selection

We tested each feature incrementally using pooled OLS on non-overlapping monthly observations.

**Kept:** reversal (t = -9.59, R² +0.2pp), mom_3m, mom_6m (incremental momentum signals)

**Dropped:**
- **Volatility** (t = -3.09): significant but overlaps with VIX — no R² improvement
- **52-week high ratio** (t = -3.69): wrong sign — reversal effect dominates at 21-day horizon
- **Beta, idiosyncratic vol**: added noise, worsened out-of-sample performance

## 3. Fama-MacBeth Regression

```python
# ── Fama-MacBeth: cross-sectional OLS each month ─────────────────────
# Addresses cross-sectional correlation ignored by pooled OLS

features_fm = ['momentum', 'reversal', 'mom_3m', 'mom_6m',
               'VIX', 'Mkt_RF', 'SMB', 'HML', 'FEDFUNDS', 'CPI_YoY', 'UNRATE']

panel_monthly = (
    panel.groupby('ticker', group_keys=False)
    .apply(lambda x: x.iloc[::21])
    .reset_index(drop=True)
)

monthly_coefs = []
for date, group in panel_monthly.groupby('date'):
    if len(group) < 10:
        continue
    X = sm.add_constant(group[features_fm])
    y = group['forward_return']
    result = sm.OLS(y, X).fit()
    monthly_coefs.append(result.params)

coefs = pd.DataFrame(monthly_coefs)
means = coefs.mean()
stds  = coefs.std()
tstat = means / (stds / np.sqrt(len(coefs)))

fm_results = pd.DataFrame({'mean_coef': means, 'std': stds, 't_stat': tstat})
print(f'Number of monthly regressions: {len(coefs)}')
print(f'\nFama-MacBeth Results:')
print(fm_results.round(6))
```

### Fama-MacBeth Findings

Only **CPI_YoY** (t = -3.28) clears the |t| > 2 threshold. Pooled OLS inflated t-statistics by ignoring cross-sectional correlation — stocks in the same month move together and are not independent observations.

The low t-stats indicate most signals are inconsistent over time, motivating the move to non-linear models.

## 4. Model Comparison — Raw Returns

```python
# ── Train/test split by time ─────────────────────────────────────────
# 80% train, 20% test — never shuffle financial data
features_all = ['momentum', 'reversal', 'mom_3m', 'mom_6m',
                'VIX', 'Mkt_RF', 'SMB', 'HML', 'FEDFUNDS', 'CPI_YoY', 'UNRATE']

features_stock = ['momentum', 'reversal', 'mom_3m', 'mom_6m']

dates = sorted(panel_monthly['date'].unique())
cutoff = dates[int(len(dates) * 0.8)]

train = panel_monthly[panel_monthly['date'] <= cutoff]
test  = panel_monthly[panel_monthly['date'] > cutoff]

X_train_all = train[features_all]
X_test_all  = test[features_all]
X_train_s   = train[features_stock]
X_test_s    = test[features_stock]
y_train     = train['forward_return']
y_test      = test['forward_return']

print(f'Train: {train["date"].min().date()} to {train["date"].max().date()} ({len(train)} rows)')
print(f'Test:  {test["date"].min().date()} to {test["date"].max().date()} ({len(test)} rows)')

# ── All features ──────────────────────────────────────────────────────
ols_all = sm.OLS(y_train, sm.add_constant(X_train_all)).fit()
print(f'\n── All features ──')
print(f'OLS   R² train: {r2_score(y_train, ols_all.predict(sm.add_constant(X_train_all))):.4f}  test: {r2_score(y_test, ols_all.predict(sm.add_constant(X_test_all))):.4f}')

ridge_all = RidgeCV(alphas=[0.001, 0.01, 0.1, 1.0, 10.0], cv=5)
ridge_all.fit(X_train_all, y_train)
print(f'Ridge R² train: {r2_score(y_train, ridge_all.predict(X_train_all)):.4f}  test: {r2_score(y_test, ridge_all.predict(X_test_all)):.4f}')

lasso_all = LassoCV(alphas=[0.0001, 0.001, 0.01, 0.1], cv=5, max_iter=5000)
lasso_all.fit(X_train_all, y_train)
print(f'Lasso R² train: {r2_score(y_train, lasso_all.predict(X_train_all)):.4f}  test: {r2_score(y_test, lasso_all.predict(X_test_all)):.4f}')

rf_all = RandomForestRegressor(n_estimators=100, max_depth=3, min_samples_leaf=200, max_features=0.5, random_state=42, n_jobs=-1)
rf_all.fit(X_train_all, y_train)
print(f'RF    R² train: {r2_score(y_train, rf_all.predict(X_train_all)):.4f}  test: {r2_score(y_test, rf_all.predict(X_test_all)):.4f}')

xgb_all = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.05, subsample=0.8, colsample_bytree=0.8, min_child_weight=50, random_state=42, n_jobs=-1)
xgb_all.fit(X_train_all, y_train)
print(f'XGB   R² train: {r2_score(y_train, xgb_all.predict(X_train_all)):.4f}  test: {r2_score(y_test, xgb_all.predict(X_test_all)):.4f}')

# ── Stock features only ───────────────────────────────────────────────
ols_s = sm.OLS(y_train, sm.add_constant(X_train_s)).fit()
print(f'\n── Stock features only ──')
print(f'OLS   R² train: {r2_score(y_train, ols_s.predict(sm.add_constant(X_train_s))):.4f}  test: {r2_score(y_test, ols_s.predict(sm.add_constant(X_test_s))):.4f}')

rf_s = RandomForestRegressor(n_estimators=100, max_depth=3, min_samples_leaf=200, max_features=0.5, random_state=42, n_jobs=-1)
rf_s.fit(X_train_s, y_train)
print(f'RF    R² train: {r2_score(y_train, rf_s.predict(X_train_s)):.4f}  test: {r2_score(y_test, rf_s.predict(X_test_s)):.4f}')

xgb_s = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.05, subsample=0.8, colsample_bytree=0.8, min_child_weight=50, random_state=42, n_jobs=-1)
xgb_s.fit(X_train_s, y_train)
print(f'XGB   R² train: {r2_score(y_train, xgb_s.predict(X_train_s)):.4f}  test: {r2_score(y_test, xgb_s.predict(X_test_s)):.4f}')
```

### Model Comparison Findings

All models produce negative out-of-sample R² — the 2022-2026 test period represents a fundamentally different macro regime (aggressive Fed tightening, CPI at 9%) versus the 2010-2022 training period.

**Key findings:**
- Removing macro features improved test R² across all models — macro signals are regime-specific
- More complex models overfit more — XGBoost has the worst test R² despite best train R²
- Linear models are more robust to regime changes when features are macro-dominated

## 5. Market-Adjusted Target (Excess Returns)

```python
# ── Compute market-adjusted target ───────────────────────────────────
spy = yf.download('SPY', start='2010-01-01', end='2026-05-16', auto_adjust=True)['Close'].squeeze()
spy_21d = spy.pct_change(21).shift(-21)
spy_21d.name = 'spy_21d'

panel_adj = panel.copy()
panel_adj = panel_adj.merge(spy_21d.reset_index().rename(columns={'Date':'date'}), on='date', how='left')
panel_adj['excess_return'] = panel_adj['forward_return'] - panel_adj['spy_21d']
panel_adj = panel_adj.dropna(subset=['excess_return'])

# ── Add volatility and 52w ratio back (behave differently vs raw target)
panel_adj = panel_adj.dropna(subset=['volatility', 'ratio_52w'])

# ── Rebuild monthly panel ─────────────────────────────────────────────
panel_monthly_adj = (
    panel_adj.groupby('ticker', group_keys=False)
    .apply(lambda x: x.iloc[::21])
    .reset_index(drop=True)
)

features_exp = ['momentum', 'reversal', 'mom_3m', 'mom_6m', 'volatility', 'ratio_52w']

dates_adj = sorted(panel_monthly_adj['date'].unique())
cutoff_adj = dates_adj[int(len(dates_adj) * 0.8)]

train_adj = panel_monthly_adj[panel_monthly_adj['date'] <= cutoff_adj]
test_adj  = panel_monthly_adj[panel_monthly_adj['date'] > cutoff_adj]

X_train_adj = train_adj[features_exp]
X_test_adj  = test_adj[features_exp]
y_train_adj = train_adj['excess_return']
y_test_adj  = test_adj['excess_return']

print(f'Train: {train_adj["date"].min().date()} to {train_adj["date"].max().date()} ({len(train_adj)} rows)')
print(f'Test:  {test_adj["date"].min().date()} to {test_adj["date"].max().date()} ({len(test_adj)} rows)')
```

```python
# ── Train all models on excess return target ──────────────────────────
ols_adj = sm.OLS(y_train_adj, sm.add_constant(X_train_adj)).fit()
print(f'OLS     R² train: {r2_score(y_train_adj, ols_adj.predict(sm.add_constant(X_train_adj))):.4f}  test: {r2_score(y_test_adj, ols_adj.predict(sm.add_constant(X_test_adj))):.4f}')

ridge_adj = RidgeCV(alphas=[0.001, 0.01, 0.1, 1.0, 10.0], cv=5)
ridge_adj.fit(X_train_adj, y_train_adj)
print(f'Ridge   R² train: {r2_score(y_train_adj, ridge_adj.predict(X_train_adj)):.4f}  test: {r2_score(y_test_adj, ridge_adj.predict(X_test_adj)):.4f}')

lasso_adj = LassoCV(alphas=[0.0001, 0.001, 0.01, 0.1], cv=5, max_iter=5000)
lasso_adj.fit(X_train_adj, y_train_adj)
print(f'Lasso   R² train: {r2_score(y_train_adj, lasso_adj.predict(X_train_adj)):.4f}  test: {r2_score(y_test_adj, lasso_adj.predict(X_test_adj)):.4f}')

rf_adj = RandomForestRegressor(n_estimators=100, max_depth=3, min_samples_leaf=200, max_features=0.5, random_state=42, n_jobs=-1)
rf_adj.fit(X_train_adj, y_train_adj)
print(f'RF      R² train: {r2_score(y_train_adj, rf_adj.predict(X_train_adj)):.4f}  test: {r2_score(y_test_adj, rf_adj.predict(X_test_adj)):.4f}')

xgb_adj = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.05, subsample=0.8, colsample_bytree=0.8, min_child_weight=50, random_state=42, n_jobs=-1)
xgb_adj.fit(X_train_adj, y_train_adj)
print(f'XGBoost R² train: {r2_score(y_train_adj, xgb_adj.predict(X_train_adj)):.4f}  test: {r2_score(y_test_adj, xgb_adj.predict(X_test_adj)):.4f}')
```

## 6. Portfolio Analytics

```python
# ── Analytics function ───────────────────────────────────────────────
def compute_analytics(X, y, dates_col, model, model_name='Model'):
    try:
        pred = model.predict(sm.add_constant(X))
    except:
        pred = model.predict(X)

    df = pd.DataFrame({
        'date'           : dates_col.values,
        'forward_return' : y.values,
        'predicted'      : pred
    })

    monthly_returns = []
    for date, group in df.groupby('date'):
        if len(group) < 10:
            continue
        q20 = group['predicted'].quantile(0.2)
        q80 = group['predicted'].quantile(0.8)
        long_ret  = group[group['predicted'] >= q80]['forward_return'].mean()
        short_ret = group[group['predicted'] <= q20]['forward_return'].mean()
        monthly_returns.append(long_ret - short_ret)

    monthly_returns = pd.Series(monthly_returns)
    sharpe     = monthly_returns.mean() / monthly_returns.std() * np.sqrt(12)
    ann_return = monthly_returns.mean() * 12
    total_ret  = (1 + monthly_returns).prod() - 1
    win_rate   = (monthly_returns > 0).mean()
    max_dd     = (monthly_returns.cumsum() - monthly_returns.cumsum().cummax()).min()

    print(f'── {model_name} ──────────────────────────')
    print(f'Sharpe ratio:      {sharpe:.4f}')
    print(f'Annualized return: {ann_return*100:.2f}%')
    print(f'Total return:      {total_ret*100:.2f}%')
    print(f'Win rate:          {win_rate*100:.1f}%')
    print(f'Max drawdown:      {max_dd*100:.2f}%')
    print()
    return monthly_returns

ret_ols   = compute_analytics(X_test_adj, y_test_adj, test_adj['date'], ols_adj,   'OLS')
ret_ridge = compute_analytics(X_test_adj, y_test_adj, test_adj['date'], ridge_adj, 'Ridge')
ret_lasso = compute_analytics(X_test_adj, y_test_adj, test_adj['date'], lasso_adj, 'Lasso')
ret_rf    = compute_analytics(X_test_adj, y_test_adj, test_adj['date'], rf_adj,    'Random Forest')
ret_xgb   = compute_analytics(X_test_adj, y_test_adj, test_adj['date'], xgb_adj,  'XGBoost')
```

## 7. Transaction Cost Analysis

Each month we rebalance the portfolio — stocks enter and exit the top/bottom quintiles. We model one-way transaction costs as a percentage of traded value.

Assumptions:
- Monthly rebalancing
- ~20% turnover per month (stocks entering/leaving quintiles)
- One-way cost applied to both long and short legs

```python
# ── Transaction cost impact on best model (XGBoost) ──────────────────
# Costs in basis points (1bp = 0.01%)
cost_levels = [0, 5, 10, 20, 50]
turnover = 0.20  # estimated 20% monthly turnover

print('Transaction Cost Analysis — XGBoost (best model)')
print(f'{"Cost (bps)":>12} {"Net Sharpe":>12} {"Net Ann Return":>16} {"Net Total Return":>18}')
print('-' * 62)

for bps in cost_levels:
    cost_per_month = (bps / 10000) * turnover * 2  # both legs
    net_returns = ret_xgb - cost_per_month
    net_sharpe  = net_returns.mean() / net_returns.std() * np.sqrt(12)
    net_ann     = net_returns.mean() * 12
    net_total   = (1 + net_returns).prod() - 1
    print(f'{bps:>12} {net_sharpe:>12.4f} {net_ann*100:>15.2f}% {net_total*100:>17.2f}%')
```

The strategy remains profitable even at 50bp one-way cost — well above typical large-cap transaction costs of 5-10bp.

### Key Findings

**1. Market-adjusted target is critical.** Predicting excess returns over SPY instead of raw returns removes market direction from the target, dramatically improving out-of-sample generalisation. The model focuses on *which stocks beat the market* rather than *where the market is going*.

**2. Macro features cause regime overfitting.** VIX, Mkt_RF, SMB, HML, FEDFUNDS, CPI_YoY, and UNRATE learned relationships in the 2010-2022 low-rate environment that completely reversed in 2022-2026. Removing them improved all models.

**3. Non-linear models outperform on stock-specific features.** Volatility and 52-week high ratio have non-linear relationships with excess returns that OLS cannot capture. XGBoost (Sharpe 0.98) and Random Forest (Sharpe 0.86) significantly outperform linear models (OLS Sharpe 0.46).

**4. Win rate is a misleading metric.** XGBoost wins 61.3% of months with the lowest max drawdown (-11.63%) — best risk-adjusted performance across all metrics.

**5. Transaction costs are manageable.** At 10bp one-way cost (realistic for large-cap S&P 500 stocks), XGBoost Sharpe drops from 0.98 to 0.93 — still excellent.

### Limitations
- Price-only features limit predictive power — fundamentals, earnings surprises, and analyst revisions would improve the model
- Survivorship bias from 26% failed yfinance downloads remains a data limitation
- Models should be retrained on a rolling basis to adapt to regime changes
- The clean train/test split used here is still a single historical episode — performance may vary across different periods
